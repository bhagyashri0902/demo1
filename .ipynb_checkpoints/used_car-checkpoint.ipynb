```python
import pandas as pd
```

```python
df=pd.read_csv('../../datasets/used_car_dataset.csv')
```

```python
df
```

```python
df['brand_enc']=df['Brand'].map({'Ambassador':0 , 'Ashok': 1,'Aston Martin':2,'Audi':3,'Bajaj':4,'Bentley':5,'BMW':6,'Chevrolet':7,'Datsun':8,'Fiat':9,
'Force':10,'Ford':11,'Honda':12,'Hyundai':13,'ICML':14,'Isuzu':15,'Jaguar':16,'Jeep':17,'Kia':18,'Land Rover':19,'Lexus':20,'Mahindra':21,'Maruti Suzuki':22,'Maserati':23,'Mercedes-Benz':24,
   'MG':25,'Mini':26,'Mitsubishi':27,'Nissan':28,'Opel':29,'Porsche':30,'Renault':31,'Rolls-Royce':32,'Skoda':33,'Ssangyong':34,'Tata':35,'Toyota':36,'Volkswagen':37,'Volvo':38})
```

```python
df
```

```python
df['Kms_enc'] = df['kmDriven'].str.replace(',', '')
df['Kms_enc']=df['Kms_enc'].str.replace('km','')
df['Kms_enc'] = df['Kms_enc'].astype(float)
df
```

```python
df['Transmission_enc']=df['Transmission'].map({'Manual':0,'Automatic':1})
df
```

```python
df['Owner_enc']=df['Owner'].map({'first':0,'second':1})
df['Fuel_enc']=df['FuelType'].map({'Petrol':0,'Diesel':1,'Hybrid/CNG':2})
df
```

```python
df['Ask_Price']=df['AskPrice'].str.replace('₹','')
df['Ask_Price']=df['Ask_Price'].str.replace(',','')
df
```

```python
X=df[['brand_enc','Year','Kms_enc','Transmission_enc','Owner_enc','Fuel_enc']]
y=df['Ask_Price']
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression() # Empty model
model.fit(X,y) # Find best fit line/train the model
```

```python
X_test=[[13,2021,83228.0,0,1,0]]
model.predict(X_test)[0]
```

```python
import pickle
```

```python
pickle.dump(model,open('used_cars.pkl','wb'))
```
